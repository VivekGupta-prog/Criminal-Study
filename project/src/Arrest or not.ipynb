```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import dataset
df= pd.read_csv('Chicago_Crimes_2015_to_2017.csv')
```

```python
df = df.dropna()    #drop null rows
```

```python
df.isnull().any()
```

```python
# Remove irrelevant/not meaningfull attributes
df = df.drop(['Unnamed: 0'], axis=1)
df = df.drop(['ID'], axis=1)
df = df.drop(['Case Number'], axis=1) 
```

```python
df
```

```python
df['Date']= pd.to_datetime(df['Date'])
```

```python
# devide dates into different columns
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Hour'] = df['Date'].dt.hour
df['Minute'] = df['Date'].dt.minute
df['Second'] = df['Date'].dt.second
```

```python
df
```

```python
# drop Date and Updated on attribute
df = df.drop(['Date'], axis=1)
df = df.drop(['Updated On'], axis=1)
```

```python
df.nunique()
```

```python
# Convert Categorical Attributes to Numerical
df['Block'] = pd.factorize(df["Block"])[0]
df['IUCR'] = pd.factorize(df["IUCR"])[0]
df['Primary Type'] = pd.factorize(df["Primary Type"])[0]
df['Description'] = pd.factorize(df["Description"])[0]
df['Location Description'] = pd.factorize(df["Location Description"])[0]
df['FBI Code'] = pd.factorize(df["FBI Code"])[0]
df['Location'] = pd.factorize(df["Location"])[0] 
```

```python
#change true and false by number to plot the data in a map with different colors
df= df.replace({False: 0, True: 1})
```

```python
#Split dataset to Training Set & Test Set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

# feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#save the model untill feature scaling 
from joblib import dump  
dump(sc,"scal.save")                     # for scaler file
```

# logistic regression

```python
from sklearn.linear_model import LogisticRegression
# Create Model with configuration
lrc = LogisticRegression()
#model training
lrc.fit(x_train,y_train)
```

```python
#prediction
y_pred_lrc = lrc.predict(x_test)
y_pred_lrc
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
```

```python
ac_lrc = accuracy_score(y_test,y_pred_lrc)
rc_lrc = recall_score(y_test,y_pred_lrc, average="weighted")
pr_lrc = precision_score(y_test,y_pred_lrc, average="weighted")
f1_lrc = f1_score(y_test,y_pred_lrc, average='micro')
cm_lrc = confusion_matrix(y_test,y_pred_lrc)

print("========== Logistic Regression Results ==========")
print("Accuracy    : ", ac_lrc)
print("Recall      : ", rc_lrc)
print("Precision   : ", pr_lrc)
print("F1 Score    : ", f1_lrc)
print("Confusion Matrix: ")
print(cm_lrc)
```

```python
import sklearn.metrics as metrics
fpr,tpr,threshold=metrics.roc_curve(y_test,y_pred_lrc)
roc_auc_lrc= metrics.auc(fpr,tpr)
roc_auc_lrc
```

```python
plt.title("Logistic Regression:ROC")
plt.plot(fpr,tpr,'b',label='AUC =%0.2f'%roc_auc_lrc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.ylabel("True +ve Rate (Sensitivity)")
plt.xlabel("False +ve Rate (1- Specificity)")
plt.show()
```
